** How many Job Titles were represented by only one person in 2013? (e.g. Job Titles with only one occurence in 2013?) **

```python
sal[sal['Year']==2013]
```

```python
sal[sal['Year']==2013].groupby('JobTitle').count()
```

```python
sal[sal['Year']==2013].groupby('JobTitle').count().groupby('Id').count()
```

```python
sal[sal['Year']==2013].groupby('JobTitle').count().groupby('Id').count().iloc[[0],[1]]
```

```python
sum(sal[sal['Year']==2013].groupby('JobTitle').count()['Id']==1)
```

```python
sum(sal[sal['Year']==2013]['JobTitle'].value_counts()==1)
```
